## Data gathering

Using Genius.com API we can build a data set of every song by a artist

```python
!pip install lyricsgenius
```

```python
import lyricsgenius
genius = lyricsgenius.Genius("genius key")
```

```python
import csv
with open('data/rappers.txt') as rappers:
    with open('data/lyrics.csv', mode='w') as csv_file:
        fieldnames = ['rapper', 'lyrics']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        for rapper in rappers:
      
            writer.writeheader()
            for rapper in rappers:
                try:
                    artist = genius.search_artist(rapper, max_songs=1000, sort="title", include_features=True)
                    
                    for song in artist.songs:
                        writer.writerow({'rapper': artist.name, 'lyrics': ", ".join(list(filter(lambda x:not ('[' in x or ']' in x),song.lyrics.split('\n'))))})
                except Exception as e:
                    print(e)
                  
```

## Building model
The simpletransformers makes building NLP model extremely easy!

We'll use the gpt2 model and train it from scratch

```python
!pip install simpletransformers
```

we'll load the data and split it to train and test sets

```python
import pandas as pd
lyrics = pd.read_csv('data/lyrics.csv')
from sklearn.model_selection import train_test_split
# test = lyrics[lyrics['rapper'].str.contains('רביד')]
train, test = train_test_split(lyrics, test_size=0.1)
```

save our data to files

```python
train['lyrics'].to_csv('/content/drive/My Drive/train.txt', sep='\n', index=False)
```

```python
test['lyrics'].to_csv('/content/drive/My Drive/test.txt', sep='\n', index=False)
```

now we'll train our model

```python
from simpletransformers.language_modeling import LanguageModelingModel
import logging


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)


train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "num_train_epochs": 20,
    "save_eval_checkpoints": True,
    "block_size": 509,
    "max_seq_length": 509,
    # "save_model_every_epoch": False,
    "learning_rate": 1e-4,
    "train_batch_size": 9,
    "gradient_accumulation_steps": 4,
    "mlm": False,
    "dataset_type": "simple",
    "logging_steps": 100,
    'use_early_stopping':True,
    "evaluate_during_training": True,
    "evaluate_during_training_steps": 3000,
    "evaluate_during_training_verbose": True,
    "use_cached_eval_features": True,
    "sliding_window": True,
    "use_multiprocessing": True,
    "vocab_size": 10000,
    "output_dir": f"/content/drive/My Drive/outputs/from_scratch",
    "best_model_dir": f"/content/drive/My Drive/outputs/from_scratch/best_model"
}

train_file = f"/content/drive/My Drive/train.txt"
# test_file = f"data/test.txt"
test_file = f"/content/drive/My Drive/test.txt"
model = LanguageModelingModel(
    "gpt2",
    None,
    # "outputs/best_model",
    # "outputs/checkpoint-28000",
    args=train_args,
    train_files=train_file,
    #use_cuda=False,
)

# model.train_tokenizer(train_file)

model.train_model(
    train_file,
    eval_file=test_file,
)

```

Now we can load our model into language generator and generate rap in hebrew!

```python
%time
from simpletransformers.language_generation import LanguageGenerationModel
model = LanguageGenerationModel("gpt2", "/content/drive/My Drive/outputs/from_scratch/best_model/", args={"max_length": 500, "temperature":1.1})


# generatedc = '.'.join(generated[0].split('.')[:-1]) + '.'
```
